```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
with open("data/sp500_earnings.csv") as f:
    sp500 = pd.read_csv(f)
```

```python
share_price = sp500[["QUARTER END", "PRICE"]]
share_price.rename(columns={"QUARTER END": "Date", "PRICE": "Share Price"}, inplace=True)
share_price = share_price[::-1]

share_price.plot(x="Date", y="Share Price", title="S&P 500 Share Price")
plt.xticks(np.arange(0, len(share_price), 8), rotation=45)

plt.show()
```
